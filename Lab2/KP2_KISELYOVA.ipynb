```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

<p><mark>Створення вектора</mark></p>

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

<p><mark>Побудова графіків</mark></p>

<p><mark>1. Функціональний метод</mark></p>

```python
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("График")
```

<p><mark>Для побудови декількох графіків функція subplot:</mark></p>

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
```

<p><mark>2. Об’єктно-орієнтований метод</mark></p>

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("График")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0.5,0.5,0.5])
axes2 = fig.add_axes([0.5,0.5,0.5,0.5])
axes1.plot(x, y,'r')
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("График1")
axes2.plot(x, y,'b')
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
фaxes2.set_title("График2")
```

<mark>Використання subplots</mark>

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 4)
```

```python
plt.tight_layout()
```

<mark>Додавання легенди</mark>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label ="x**2")
ax.plot(x, x**3, label ="x**3")
ax.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi =100)
```

```python
fig = plt.figure(figsize=(8,4), dpi =100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label ="x**2")
ax.plot(x, x**3, label ="x**3")
ax.legend()
```

<mark>Задати колір графіку: </mark>

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, color = 'g')
ax.plot(x, x**3, color = 'y')
```

<mark>Задати тип та ширину лінії: </mark>

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, lw = 2, ls ='--')
```

<mark>тип, розмір та колір маркерів</mark>

```python
fig, ax = plt.subplots()
ax.plot(x, x**2,marker ="o", markersize = 10,markerfacecolor = 'y')
```

<mark>Збереження графіку</mark>

```python
fig.savefig("grafik.png")
```

<h3><mark>Побудова спеціальних графіків</mark></h3>
<mark>- точкова діаграма</mark>

```python
plt.scatter(x,y)
```

<mark>- гістограма</mark>

```python
from random import sample
data = sample(range(1,1000), 100)
plt.hist(data)
```

<mark>- box-plot діаграма</mark>

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert = True, patch_artist = True)
```

<p>Это используется для визуализации отношений между X и Y вместе с одномерным распределением X и Y по отдельности. Этот график часто используется в анализе данных (EDA).</p>
