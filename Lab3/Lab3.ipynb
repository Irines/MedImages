```python
import seaborn as sns
%matplotlib inline
```

<h4>Тестові дані </h4>

```python
tips = sns.load_dataset('tips')
```

<h4>Візуалізація даних</h4>

<h4>Графіки для аналізу кількісних даних</h4>

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=5)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips)
```

<blockquote> Можливі варіанти для модифікації:
kde – за замовчуванням True. Якщо False – на графіку буде відображено
лише гістограму, без кривої розподілу.
bins – ширина «кроку» для гістограми.
</blockquote>

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='kde')
```

<blockquote>jointplot – дозволяє співставити розподіл двох змінних та їх
взаємозв’язок за допомогою діаграми розсіювання.
Можливі варіанти для модифікації:
kind – варіант діаграми розсіювання. Можливі варіанти – scatter (за
замовчуванням), reg, resid, kde, hex.
pairplot – будує попарні взаємозв’язки між змінними в повному наборі
даних.
По діагоналі відображені графіки розподілу даних, в інших комірках -
діаграми розсіювання (для відображення взаємозв’язку між двома змінними). </blockquote>   

```python
sns.pairplot(tips)
```

<blockquote>Можливі варіанти для модифікації:
hue – вибір категоріальної змінної для розділення даних на окремі
підгрупи та для їх кольорового відображення на діаграмі.
palette – кольорова схема для візуалізації
(http://seaborn.pydata.org/tutorial/aesthetics.html)</blockquote>

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='smoker',palette="deep")
```

Графіки для аналізу категоріальних даних

<blockquote>barplot – дозволяють отримувати сукупні дані по кількісним даним в
залежності від категоріальної змінної.</blockquote>

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

<blockquote>Для даного графіку х – категоріальна змінна, у – кількісна. За
замовчуванням відображено середнє значення кількісної змінної.
Можливі варіанти для модифікації:
hue – вибір додаткової категоріальної змінної для розділення даних на
окремі підгрупи.
estimator – зміна варіанту відображення кількісної змінної. Наприклад,
не середнє значення, а медіана, мода тощо. Даному параметру можна
призначити як власні функції, так і функції з різних бібліотек. Наприклад:</blockquote>

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.median)
```

<blockquote>boxplots - використовується для візуалізації розподілу категоріальних
даних за певною кількісною змінною. Кожен «ящик» демонструє
інтерквартильний розмах (25%, 50% та 75%), «вуса» демонструють повний
розкид даних за виключенням викидів (аномально високі чи аномально
низькі значення).</blockquote>

```python
sns.boxplot(x='day',y='total_bill', data=tips, palette='rainbow')
```

<blockquote>Можливі варіанти для модифікації:
hue – вибір додаткової категоріальної змінної для розділення даних на
окремі підгрупи.</blockquote>

```python
sns.boxplot(x='day',y='total_bill',hue='sex', data=tips, palette='rainbow')
```

<blockquote>stripplot – представляє собою діаграму розсіювання, де одна за змінних
категоріальна.</blockquote>

```python
sns.stripplot(x="day", y="total_bill",data=tips)
```

<blockquote>Можливі варіанти для модифікації:

hue – вибір додаткової категоріальної змінної для розділення даних на
окремі підгрупи.
dodge – додається лише за умови вибору параметру hue. За
замовчуванням False. Дозволяє розділити групи при візуалізації.</blockquote>

```python
sns.stripplot(x="day", y="total_bill",hue="sex",data=tips)
```

```python
sns.stripplot(x="day", y="total_bill",hue="sex",dodge=True,data=tips)
```

<h3>Графіки для аналізу взаємозв’язку даних</h3>

<blockquote>Heatmap – графік взаємозв’язку (кореляції) між даними у вигляді
матриці. Для побудови даного графіку необхідно попередньо розрахувати
матрицю кореляцій між змінними ( .corr).</blockquote>

```python
sns.heatmap(tips.corr())
```

Можливі варіанти для модифікації:
annot – додавання коефіцієнтів кореляції на діаграму.

<h3>Графіки для відображення даних у вигляді сітки</h3>

1. Побудуйте площу для відображення даних, обравши одну або дві
кількісні змінні.

```python
g=sns.FacetGrid(tips, col="time",row="smoker")
```

2. Додайте на площу дані для відображення:

```python
from matplotlib import pyplot as plt
g=sns.FacetGrid(tips,col="time",row="smoker")
g=g.map(plt.hist,"total_bill")
```

<blockquote>Можливі варіанти для модифікації:
hue – вибір додаткової категоріальної змінної для розділення даних на
окремі підгрупи.
.add_legend() – додавання легенди до діаграми</blockquote>

```python
g=sns.FacetGrid(tips,col="time",row="smoker",hue="sex")
g=g.map(plt.hist,"total_bill")
```

```python
bins = np.arange(0, 65, 5)
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill", bins=bins, color="r")
```

```python
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g = g.map(plt.hist, "total_bill", bins=bins)
```

```python
kws = dict(s=50, linewidth=.5, edgecolor="w")
g = sns.FacetGrid(tips, col="sex", hue="time", palette="Set1",
                  hue_order=["Dinner", "Lunch"])
g = (g.map(plt.scatter, "total_bill", "tip", **kws)
     .add_legend())
```

```python
g = sns.FacetGrid(tips, col="sex", hue="time", palette="Set1",
                  hue_order=["Dinner", "Lunch"])
g = (g.map(plt.scatter, "total_bill", "tip")
     .add_legend())
```
